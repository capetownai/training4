```python
import os
bulletins=os.listdir("data/Job Bulletins/")
```

```python
import re

# rm ._* 
# ls -a

def get_requirements(bulletin):       
    
    with open("data/Job Bulletins/"+bulletins[bulletin]) as f:    ##reading text files
        try:
            txt=f.read()
            start = 'REQUIREMENT'
            end = 'PROCESS'
            txt = txt[txt.find(start)+len(start):txt.find(end)]
            txt = txt.replace('/MINIMUM QUALIFICATION', '')
            txt = txt.replace('\n', '')
            return txt
        except:
            return ''
```

```python
requirements = ''
for i in range(0,len(bulletins)):
    requirements += get_requirements(i)
```
